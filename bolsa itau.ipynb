```python
from pandas_datareader import data as web
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import seaborn as sns
```

```python
banco_cot = web.DataReader('ITUB4.SA',data_source='yahoo',start='01-01-2022',end='01-31-2022')
banco_cot
```

```python
banco_cot['Var']=banco_cot['Close']-banco_cot['Open']
banco_cot
```

```python
plt.figure(figsize=(15,10))
plt.xlabel('Data')
plt.ylabel('Valor')
plt.title('Gráfico da ação ITUB4.SA do periodo de janeiro de 2022')
sns.lineplot(x='Date',y='Open',data=banco_cot, label='Abertura')
sns.lineplot(x='Date',y='Close',data=banco_cot,label='Fechamento')
sns.lineplot(x='Date',y='Var',data=banco_cot,label='Variação')
plt.show()
```
